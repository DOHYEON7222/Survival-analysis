# What is survival analysis?

: 생존분석이란 관심있는 사건이 발생할 때까지 예상되는 기간을 조사하는 통계학적인 방법이다.


> **생존분석의 필요한 개념**
> - *Event*: 생존 or 사망 즉, 관심있는 사건
> - *survival time*: 관심있는 사건이 발생할 때가지의 기간
---
> **Censorship problem type**
> 1. *left censored*: 생존 기간이 관찰된 시간보다 짧음 
> 2. *right censored*: 생존 기간이 관찰된 시간보다 김
> 3. *interval censored*: 생존 기간을 명확하게 정의할 수 없음
---
가장 일반적인 유형은 right censored이고, 일반적으로 생존분석에서 처리되는 데이터이다.  
나머지는 데이터가 문제가 있을 때 발생할 수 있다

# Survival function


- T는 사건이 발생하고 t가 관측 중인 임의의 시점일 때, 생존 S(t)는 T가 t보다 클 확률이다.
- 즉 생존함수는 시간 t 이후에 개인이 생존할 확률
$$S(t) = \Pr(T > t)$$


   
#### **Survival function characteristics**
    - T ≥ 0 and 0 < t < ∞
    - It is non-increasing
    - If t=0, then  (t)=1(survival probability is 1 at time 0)
    - If t=∞, then S(t)=0(survival probability goes to 0 as time goes to infinity

# Hazard function

: 개인이 시간 t까지 생존하고, 정확하게 시간 t에 관심있는 사건을 경험할 확률

$$h(t) = -\frac{d}{dt} \log S(t)$$

기본적으로 생존함수는 다음과 같이 정의될 수 있다
$$S(t) = P(T > t) = 1 - F(t)$$

위험함수의 정의는 다음과 같으며, ***생존시간이 t보다 클때 바로 사망할 확률*** 을 의미하고, 이를 **순간위험율**이라고 한다.
$$
\begin{aligned}
h(t) &= \lim_{\Delta t \to 0} \frac{P(t < T \leq t + \Delta t \mid T > t)}{\Delta t} \\
     &= \lim_{\Delta t \to 0} \frac{1}{\Delta t} \cdot \frac{P(t < T \leq t + \Delta t)}{P(T > t)} \\
     &= \frac{f(t)}{S(t)}
\end{aligned}
$$


---
#### 위험함수의 의미??
    - 위험함수는 아주 짧은 시간에 사건이 일어날 확률로써, 굉장히 순간적인 확률적인 의미를 가진다.
    - 사실, 확률보다는 rate (발생률)이라고 생각하면 되고, 위험함수의 값은 [0, 1] 사이의 값을 가지기 않는다.
    - 위 식에서 2번째 식을 보면, 전체 모든 환자에 대한 것이 아니라, 특정 시간 t까지 살아있는 모든 환자군에 대한 위험률이다.
    - 생존분석에서는 위험함수가 더 큰 의미를 가진다.
---
=> 위험함수는 시간에 따른 위험의 변화를 나타나기 대문에, 생존함수보다는 위험의 변화에 대한 정보를 잘 파악할 수 있다는 장점이 존재한다.
